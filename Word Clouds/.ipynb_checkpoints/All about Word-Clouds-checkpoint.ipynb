```python
import matplotlib.pyplot as plt
import wordcloud
import pandas as pd
```

```python
df = pd.read_csv('Youtube04-Eminem.csv')
```

```python
df
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopwords= set(STOPWORDS)
```

```python
type(df['CONTENT'][0])
```

```python
comments_words = ''
```

```python
for val in df.CONTENT:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    comments_words += " ".join(tokens) + " "
    
```

```python
# def token_split(x):    
#     for token in x.split():
#         token = token.lower()
#         comments_words += token + " "
        

# df['CONTENT'].apply(token_split)
```

```python
wordcloud = WordCloud(width=400 , height=400 , margin=2 , background_color='white' , stopwords=stopwords,min_font_size=10).generate(comments_words)
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis("off")
          
```

```python
from PIL import Image
import numpy as np
from wordcloud import ImageColorGenerator
```

```python
custom_mask = np.array(Image.open('Crab Mask.jpg'))
```

```python
wordcloud = WordCloud(width=200 , height=200 , margin=2 , background_color='white' ,stopwords=stopwords,min_font_size=6 , mask=custom_mask).generate(comments_words)
```

```python
image_colors = ImageColorGenerator(custom_mask)
wordcloud.recolor(color_func=image_colors)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()  # Basically you need to generate a mask in black and white.
# The part you want to show text in must be in black and the rest must be in white
```

```python
import re
from collections import Counter
```

```python
words = re.findall(r'\w+', open('example.txt').read().lower())
```

```python
words = re.findall(r'\w+', open('example.txt').read().lower())  # \w+ captures one or more word characters..
 # Just as to exclude punctuation and numbers from being counted in our word cloud...
```

```python
words_counter = Counter(words)   #
```

```python
wordcloud = WordCloud(background_color="white" , width=7000 , height=5500 , scale=2)
wordcloud.generate_from_frequencies(words_counter)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
