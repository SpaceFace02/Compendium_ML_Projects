```python
#Imputing values is just filling in the missing values,if there are relatively less number of null values..Most prob you can fill in the average age per class, or you can create a different model for just this..
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass ==1:
            return 37
        elif Pclass ==2:
            return 27
        else:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age' , 'Pclass']].apply(impute_age , axis=1)
```

```python
# In the categorical column ,like the sex column or the Embarked column, we can't pass it directly to the model, you need to convert it into a binary value for the model to accept..
```

```python
embark = pd.get_dummies(train['Embarked']  ,drop_first = True)  #No such problem here..ig, but drop one column just to be safe
```

```python
embark  # As you can see, the columns are not perfect predictors of each other
```

```python
train = pd.concat([train , sex ,embark ] , axis=1)# we want to add them along the columns
```
